```python
import requests as re
import json
import sys
import pymysql
from sqlalchemy import create_engine
import pandas as pd
```

### Conection to Google Cloud Database to import table 'words'

```python
driver = 'mysql+pymysql'
ip = ''
username = ''
password = ''
db = ''

cs  = f'{driver}://{username}:{password}@{ip}/{db}'
engine = create_engine(cs)
```

```python
Ssql = '''SELECT 
    word, 
    interactions, 
    appearances, 
    round((interactions)/(appearances),2) as eng_ratio 
FROM words 
ORDER by round((interactions)/(appearances),2) desc, appearances'''

words = pd.read_sql(Ssql,engine)
words = words[['word','interactions', 'appearances', 'eng_ratio']]
```
