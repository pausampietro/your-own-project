```python
from textblob import TextBlob
import pandas as pd
import pymysql
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats
import numpy as np
from googletrans import Translator
from google.cloud import translate
import emoji
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

### Conection to Google Cloud Database (to import tweets)

```python
driver = 'mysql+pymysql'
ip = ''
username = ''
password = ''
db = ''

cs  = f'{driver}://{username}:{password}@{ip}/{db}'
engine = create_engine(cs)
```

```python
Ssql = '''SELECT 
    MAX(tweets.`index`) AS `index`,
    MAX(`text`) AS `text`,
    MAX(url) AS url,
    MAX(`date`) AS `date`,
    MAX(retweets) AS retweets,
    MAX(likes) AS likes,
    MAX(tweets.media) AS media
FROM
    tweets
GROUP BY tweets.`text`'''

tweets = pd.read_sql(Ssql,engine)
```

## Data

* tweets: DataFrame with all the tweets from different sources
* media: DataFrame with aggregated values (group by media)

#### Translating tweets to apply sentiment analysis

* ALERT: this translation uses a google API, in case it gives an error (HTTP not available...) 
it's necessary to connect the pc to a dynamic IP (ex: phone internet sharing)

```python
def extract_emojis(str):
    return ''.join(c for c in str if c not in emoji.UNICODE_EMOJI)
```

```python
translate_client = translate.Client()
trans_text = tweets['text'].apply(lambda x: translate_client.translate(extract_emojis(x), target_language='en'))
```

```python
trans_text_def = [i['translatedText'] for i in trans_text]
```

### Creating tweets dataframe with both text in spanish/catalan and english

```python
tweets.insert(2, 'trans_text', trans_text_def)
```

### Adding columns with the result of the sentiment analysis (with TextBlob)

```python
tweets['global_subjectivity'] = tweets['trans_text'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
tweets['global_polarity_tb'] = tweets['trans_text'].apply(lambda x: TextBlob(x).sentiment[0])
```

### Adding columns with the result of the sentiment analysis (with Vader)

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
tweets['pos_scores'] = tweets['trans_text'].apply(lambda x: analyzer.polarity_scores(x)['pos'])
tweets['neg_scores'] = tweets['trans_text'].apply(lambda x: analyzer.polarity_scores(x)['neg'])
tweets['neu_scores'] = tweets['trans_text'].apply(lambda x: analyzer.polarity_scores(x)['neu'])

tweets['global_polarity'] = pd.Series(tweets['pos_scores']*1 + tweets['neg_scores']*(-1) + tweets['neu_scores']*(0))
```

#### Goodness of fit (normal distribution) for subjectivity and polarity

```python
stats.kstest(tweets['global_subjectivity'], 'norm')
```

```python
stats.kstest(tweets['global_polarity'], 'norm')
```

### Comparing scores of polarity and subjectivity in all the tweets

```python
plt.style.use('ggplot')
# plt.xkcd(scale=0.3)
f, ax = plt.subplots(1,2, figsize=(12,5), constrained_layout=True)
ax[1].hist(tweets['global_subjectivity'], bins=15, width=0.05 , color='turquoise')
ax[0].hist(tweets['global_polarity'], bins=15, width=0.05, color= 'turquoise')
f.suptitle('Polarity and Subjectivity scores in tweets', fontsize=20,)
ax[0].set_ylabel('number of tweets',fontsize=14)
ax[1].set_ylabel('number of tweets',fontsize=14)
ax[0].set_xlabel('polarity',fontsize=14)
ax[1].set_xlabel('subjectivity',fontsize=14)
ax[0].set_xlim(left=-1, right=1)
ax[1].set_xlim(left=0, right=1)
plt.show()
```

### Checking relationship between both polarity and subjectivity scores

```python
plt.title('Global subjectivity vs. global polarity of tweets')
sb.regplot(tweets['global_subjectivity'][(tweets['global_polarity']>0) &(tweets['global_subjectivity']>0)],tweets['global_polarity'][(tweets['global_polarity']>0) &(tweets['global_subjectivity']>0)])
```

### Mean values of polarity and subjectivity By 'Media'

```python
media = tweets[['media','global_subjectivity','global_polarity']].groupby('media').mean()
media
```

#### +Relative position of the media according to subjectivity/polarity

```python
area = tweets[['media','global_polarity']].groupby('media').count()
colors = ('b','purple','g','r','orange','brown')
plt.figure(figsize=(10,6))
for i in range(len(media.index)):
    plt.scatter(media['global_subjectivity'].iloc[i],media['global_polarity'].iloc[i], c=colors[i], marker='X', s=area*2)   
    plt.legend(media.index, loc=2, fontsize = 14)

plt.ylabel('mean polarity',fontsize=14)
plt.xlabel('mean subjectivity',fontsize=14)
plt.title('Relative position of the media according to subjectivity and polarity',fontsize=16, pad=15)
```
