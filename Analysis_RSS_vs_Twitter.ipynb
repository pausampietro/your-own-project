```python
from textblob import TextBlob
import pandas as pd
import pymysql
from sqlalchemy import create_engine
from googletrans import Translator
from google.cloud import translate
import emoji
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats
import numpy as np
```

### Conection to Google Cloud Database (to import tweets)

```python
driver = 'mysql+pymysql'
ip = ''
username = ''
password = ''
db = ''

cs  = f'{driver}://{username}:{password}@{ip}/{db}'
engine = create_engine(cs)
```

```python
Ssql = '''select 
    max(tweets.`index`) as `index`, 
    max(`text`) as `text`, 
    max(url) as url, 
    max(`date`) as `date`, 
    max(retweets) as retweets,
    max(likes) as likes,
    max(tweets.media) as media, 
    max(rss.`index`) as `index`,
    max(title) as title,
    max(description) as description,
    max(pubdate) as date_rss,
    max(category) as category,
    max(rss.media) as media_rss
from tweets 
inner join rss on url = link 
group by tweets.`text`'''

news = pd.read_sql(Ssql,engine)
```

## Data

* News: DataFrame with all the matches between RSS & Twitter
* Ctab: DataFrame with the numeric fields to analyze
* Dstats: DataFrame with the descriptive statistics for numeric columns (mean, std, etc)

### Generating DataFrame with NEWS+TWEETS

```python
news.rename(columns={'text':'tweet', 'title':'new'}, inplace = True)
```

##### Translating tweets to apply sentiment analysis

* ALERT: this translation uses a google API, in case it gives an error (HTTP not available...) 
it's necessary to connect the pc to a dynamic IP (ex: phone internet sharing)

```python
def extract_emojis(str):
    return ''.join(c for c in str if c not in emoji.UNICODE_EMOJI)
```

```python
translate_client = translate.Client()
trans_tweet = news['tweet'].apply(lambda x: translate_client.translate(extract_emojis(x), target_language='en'))
```

```python
translate_client = translate.Client()
trans_new = news['new'].apply(lambda x: translate_client.translate(extract_emojis(x), target_language='en'))
```

```python
trans_tweet = [i['translatedText'] for i in trans_tweet]
trans_new = [i['translatedText'] for i in trans_new]
news['t_tweet'] = pd.Series(trans_tweet)
news['t_new'] = pd.Series(trans_new)
```

### Adding sentiment analysis for 'New text' and 'Tweet text'

```python
analyzer = SentimentIntensityAnalyzer()
```

#### Scores for the NEW

```python
news['new_sub'] = news['t_new'].apply(lambda x: TextBlob(x).sentiment[1])
news['new_pos'] = news['t_new'].apply(lambda x: analyzer.polarity_scores(x)['pos'])
news['new_neg'] = news['t_new'].apply(lambda x: analyzer.polarity_scores(x)['neg'])
news['new_neu'] = news['t_new'].apply(lambda x: analyzer.polarity_scores(x)['neu'])

news['new_pol'] = pd.Series(news['new_pos']*1 + news['new_neg']*(-1) + news['new_neu']*(0))
```

#### Scores for the TWEET

```python
news['twe_sub'] = news['t_tweet'].apply(lambda x: TextBlob(x).sentiment[1])
news['twe_pos'] = news['t_tweet'].apply(lambda x: analyzer.polarity_scores(x)['pos'])
news['twe_neg'] = news['t_tweet'].apply(lambda x: analyzer.polarity_scores(x)['neg'])
news['twe_neu'] = news['t_tweet'].apply(lambda x: analyzer.polarity_scores(x)['neu'])

news['twe_pol'] = pd.Series(news['twe_pos']*1 + news['twe_neg']*(-1) + news['twe_neu']*(0))
```

### Generating comparison table New vs. Tweet

```python
ctab = news[['media','new','tweet','retweets','likes', 'new_sub', 'new_pol', 'new_pos', 'new_neg', 'new_neu', 'twe_sub', 'twe_pol', 'twe_pos', 'twe_neg', 'twe_neu', 'url']]
ctab.head(3)
```

### + Engagement 

```python
engagement = pd.DataFrame({'tweet': ctab['tweet'], 'media': ctab['media'], 'interactions': ctab['likes']+ctab['retweets'], 'tweet_polarity': ctab['twe_pol']})
engagement = engagement.sort_values(by='interactions', ascending=False).reset_index(drop=True)
engagement.head(10)
```

```python
eng_pos = engagement[engagement.tweet_polarity > 0]
eng_neg = engagement[engagement.tweet_polarity < 0]
eng_neu = engagement[engagement.tweet_polarity == 0]
```

```python
plt.style.use('ggplot')
# plt.xkcd(scale=0.3)
f, ax = plt.subplots(1,3, figsize=(15,7), constrained_layout=True)

sb.regplot(eng_pos['tweet_polarity'], eng_pos['interactions'], ax=ax[0])
sb.regplot(eng_neu['tweet_polarity'], eng_neu['interactions'], ax=ax[1])
sb.regplot(eng_neg['tweet_polarity'], eng_neg['interactions'], ax=ax[2])

ax[0].set_title('Engagement with positive polarity', fontsize=16)
ax[0].set_xlabel('polarity (>0)')
ax[0].set_ylabel('number of interactions')

ax[1].set_title('Engagement with neutral polarity', fontsize=16)
ax[1].set_xlabel('polarity (=0)')
ax[1].set_ylabel('number of interactions')

ax[2].set_title('Engagement with negative polarity', fontsize=16)
ax[2].set_xlabel('polarity (<0)')
ax[2].set_ylabel('number of interactions')
ax[2].set_ylim(0,60)


```
